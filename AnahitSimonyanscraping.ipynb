```python
from bs4 import BeautifulSoup
import requests

headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/116.0.0.0 Safari/537.36"
}

url = "https://www.imdb.com/chart/top/?ref_=nv_mv_250"
resp = requests.get(url, headers=headers)
print(resp.status_code)
```

```python
soup = BeautifulSoup(resp.content, parser = "html.parser")
```

###Title

```python
soup.find("a", class_ = "ipc-title-link-wrapper").text
```

###Rank

```python
soup.find("a", class_ = "ipc-title-link-wrapper").text[0]
```

###Year

```python
soup.find("span", class_="sc-5bc66c50-6 OOdsw cli-title-metadata-item").text
```

###Age rating

###Star rating

```python
soup.find("span", class_ = "ipc-rating-star--rating").text
```

###Number of voters

```python
soup.find("span", class_="ipc-rating-star--voteCount").text
```

```python
director_name = []

for i in soup.find_all("li role", class_="ipc-inline-list__item"):
  k = i.find("a", class_ = "ipc-metadata-list-item__list-content-item ipc-metadata-list-item__list-content-item--link")["class"][1]
  director_name.append(k)

print(director_name)
```
