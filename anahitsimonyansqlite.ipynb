```python
import sqlite3
import pandas as pd

connection = sqlite3.connect("tips.db")
cursor = connection.cursor()
```

```python
query1 = """
SELECT *
FROM Observation
WHERE total_bill > 50;
"""
pd.read_sql(query1, connection)
```

```python
query2 = """
SELECT AVG(tip) AS average_tip
FROM Observation;
"""
pd.read_sql(query2, connection)
```

```python
query3 = """
SELECT MAX(total_bill) AS max_total_bill
FROM Observation;
"""
pd.read_sql(query3, connection)
```

```python
query4 = """
SELECT Day.day, COUNT(Observation.size) AS total_people_served
FROM Observation
JOIN Day ON Observation.day_id = Day.day_id
GROUP BY Day.day;
"""
pd.read_sql(query4, connection)
```

# **<font color='red'> For counting the number of served people you should have used SUM function on Size. Minus 1 point </font>**

```python
query5 = """
SELECT size, AVG(total_bill) AS avg_total_bill
FROM Observation
GROUP BY size;
"""
pd.read_sql(query5, connection)
```

```python
query6 = """
SELECT Smoker.smoker, SUM(Observation.tip) AS total_tip
FROM Observation
JOIN Smoker ON Observation.smoker_id = Smoker.smoker_id
GROUP BY Smoker.smoker;
"""
pd.read_sql(query6, connection)
```

```python
query7 = """
SELECT Day.day, AVG(tip / total_bill * 100) AS avg_tip_percentage
FROM Observation
JOIN Day ON Observation.day_id = Day.day_id
GROUP BY Day.day;
"""
pd.read_sql(query7, connection)
```

```python
query8 = """
SELECT Observation.total_bill, Day.day, Time.time
FROM Observation
JOIN Day ON Observation.day_id = Day.day_id
JOIN Time ON Observation.time_id = Time.time_id;
"""
pd.read_sql(query8, connection)
```

```python
query9 = """
SELECT Sex.sex, COUNT(Observation.sex_id) AS total_observations
FROM Observation
JOIN Sex ON Observation.sex_id = Sex.sex_id
GROUP BY Sex.sex;
"""
pd.read_sql(query9, connection)
```

```python
query10 = """
SELECT Observation.tip
FROM Observation
JOIN Day ON Observation.day_id = Day.day_id
JOIN Time ON Observation.time_id = Time.time_id
WHERE Day.day = 'Sat' AND Time.time = 'Dinner';
"""
pd.read_sql(query10, connection)
```

```python
query11 = """
SELECT Sex.sex, MAX(Observation.total_bill) AS max_total_bill
FROM Observation
JOIN Sex ON Observation.sex_id = Sex.sex_id
GROUP BY Sex.sex;
"""
pd.read_sql(query11, connection)
```

```python
query12 = """
SELECT AVG(Observation.tip) AS avg_tip
FROM Observation
WHERE Observation.size > 4;
"""
pd.read_sql(query12, connection)
```

```python
query13 = """
SELECT Observation.*
FROM Observation
JOIN Day ON Observation.day_id = Day.day_id
JOIN Time ON Observation.time_id = Time.time_id
WHERE Day.day IN ('Fri', 'Sat') AND Time.time IN ('Lunch', 'Dinner');
"""
pd.read_sql(query13, connection)
```

```python
query14 = """
SELECT Observation.*
FROM Observation
WHERE (Observation.tip / Observation.total_bill) * 100 > 20;
"""
pd.read_sql(query14, connection)
```

```python
query15 = """
SELECT Observation.*
FROM Observation
WHERE Observation.tip = (
    SELECT MAX(Observation.tip)
    FROM Observation
    WHERE Observation.tip < (SELECT MAX(Observation.tip) FROM Observation)
);
"""
pd.read_sql(query15, connection)
```
